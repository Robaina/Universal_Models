## Add EC numbers to model reactions

Ok, looks like the universal prok model I made doesn't have EC numbers assigned to reactions. This is easy to fix since modelSEED (reactions.json) provides a mapping between reaction IDs (Bigg IDs) and EC numbers through the aliases field.

```python
import json
import cobra

# Load the JSON file
with open('reactions.json', 'r') as file:
    reaction_data = json.load(file)

# Create a dictionary to map BiGG IDs to EC numbers
bigg_to_ec = {}
for reaction in reaction_data:
    if "aliases" in reaction and reaction["aliases"] is not None:
        # Extract BiGG IDs from the aliases list
        bigg_ids = [alias.split(":")[1].strip() for alias in reaction["aliases"] if alias.startswith("BiGG:")]
        for bigg_id in bigg_ids:
            ec_numbers = reaction.get("ec_numbers", [])
            if ec_numbers:
                bigg_to_ec[bigg_id] = ec_numbers

# Assign EC numbers to reactions based on their BiGG IDs
for reaction in gem.reactions:
    if reaction.id in bigg_to_ec:
        if bigg_to_ec[reaction.id]:
            reaction.annotation['ec_number'] = bigg_to_ec[reaction.id]
    else:
        reaction.annotation['ec_number'] = [""]

# Optionally, save the modified model
cobra.io.write_sbml_model(gem, "models/universal_prokaryote_curated_ec_numbers.xml")
```

Testing assigned EC numbers, let's retrieve reaction objects by EC numbers

```python
# Get reactions by ec number
def get_reactions_by_ec_number(model, ec_numbers):
    matched_reactions = []
    for reaction in model.reactions:
        # Check if the reaction has 'ec_number' annotation and if any of the specified EC numbers are in it
        if 'ec_number' in reaction.annotation:
            if any(ec in reaction.annotation['ec_number'] for ec in ec_numbers):
                matched_reactions.append(reaction)
    return matched_reactions


ec_numbers = ['2.3.1.40','1.3.3.6','3.5.1.16']
target_rxns = get_reactions_by_ec_number(gem, ec_numbers)
```

```python
for reaction in target_rxns:
    print(f"Reaction ID: {reaction.id}, Name: {reaction.name}, EC Numbers: {reaction.annotation.get('ec_number')}")
```

## Creating the incidence matrix of the metabolic bipartite graph

We will use  the stoichiometric matrix S, the incidence matrix of the metabilic hypergraph, to transform it into a regular bipartite graph by addding a new class of nodes (reactions), which are the edges of the hypergraph.

```python
from cobra.util import create_stoichiometric_matrix

S = create_stoichiometric_matrix(gem, array_type='DataFrame')
S
```

### Node classes

Here are the two node classes of the bipartite graph: reactions (enzymes) and metabolites (compounds)

```python
reaction_nodes = {rxn.id: rxn.name for rxn in gem.reactions}
metabolite_nodes = {m.id: m.name for m in gem.metabolites}
```

### Building a directed bipartite graph representation of the metabolic network

Let's now use the stoichiometric matrix to build a directed bipartite graph with reaction and metabolite node classes. We will build the incidence and adjacency matrices of the bipartite graph as well.

```python
import networkx as nx

# Extract metabolite and reaction IDs
metabolites = [met.id for met in gem.metabolites]
reactions = [rxn.id for rxn in gem.reactions]

B = nx.DiGraph()

# Add nodes
B.add_nodes_from(metabolites, bipartite=0)
for reaction in gem.reactions:
    ec_number = reaction.annotation.get("ec_number", [""])[0] # use only first ec number
    B.add_node(reaction.id, bipartite=1, ec_number=ec_number)

# Add edges to Bipartite Graph
for row in S.itertuples(index=True):
    metabolite = row.Index
    for i, coef in enumerate(row[1:], 1):
        if coef < 0:
            B.add_edge(metabolite, S.columns[i-1])  # metabolite to reaction
        elif coef > 0:
            B.add_edge(S.columns[i-1], metabolite)  # reaction to metabolite
```

## Visualize a sample of the bipartite graph

Displaying reaction nodes in blue and metabolite nodes in pink.

```python
def visualize_sampled_subgraph(B, metabolites, reactions, num_nodes=10, display_ec_numbers=True, seed=42):
    """
    Visualizes a sampled subgraph of the bipartite graph B.

    Parameters:
    B (nx.DiGraph): The bipartite graph.
    metabolites (list): List of metabolite nodes.
    reactions (list): List of reaction nodes.
    num_nodes (int): Number of nodes to sample for visualization.
    display_ec_numbers (bool): Flag to display reaction IDs or EC numbers.
    """
    # Sample edges
    if display_ec_numbers:
        edges = [(u, v) for u, v in B.edges if B.nodes[v].get("ec_number", "") != "" or B.nodes[u].get("ec_number", "") != ""]
    else:
        edges = list(B.edges)
    random.seed(seed)
    sampled_edges = random.sample(edges, min(num_nodes, len(edges)))

    # Create a subgraph from the sampled edges
    subgraph_nodes = set()
    for u, v in sampled_edges:
        subgraph_nodes.add(u)
        subgraph_nodes.add(v)
    subgraph = B.subgraph(subgraph_nodes)

    # Prepare labels for visualization
    labels = {}
    for node in subgraph.nodes:
        if node in reactions:
            labels[node] = subgraph.nodes[node]["ec_number"] if display_ec_numbers else node
        else:
            labels[node] = node

    # Visualize the sampled subgraph
    pos = {}
    subgraph_metabolites = [node for node in subgraph.nodes if node in metabolites]
    subgraph_reactions = [node for node in subgraph.nodes if node in reactions]
    pos.update((node, (1, index)) for index, node in enumerate(subgraph_metabolites))  # Metabolites on the left
    pos.update((node, (2, index)) for index, node in enumerate(subgraph_reactions))    # Reactions on the right

    plt.figure(figsize=(10, 10))
    colors = ['pink' if n in metabolites else 'lightblue' for n in subgraph.nodes()]
    nx.draw(
        subgraph, pos, with_labels=True, labels=labels, node_size=3000,
        font_size=8, edge_color='gray', arrowstyle='-|>', arrowsize=10, node_color=colors
    )
    plt.title("Sampled Directed Bipartite Graph of Reactions and Metabolites")
    plt.show()
```
